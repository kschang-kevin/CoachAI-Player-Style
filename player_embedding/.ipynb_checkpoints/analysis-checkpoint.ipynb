```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import os
from scipy.stats import lognorm, entropy
import seaborn as sn
from itertools import product
```

```python
data = pd.read_csv('aug_set.csv')
```

```python
for rally in data['rally'].unique():
    if data.loc[data['rally']==rally]['flaw'].any():
        data = data.loc[data['rally']!=rally]

for col in data.columns:
    print(col, data[col].isnull().sum())
```

## Prelimanary Analysis
* Some simple statistic and visualization for understanding the data

```python
rally_length = [len(data.loc[data['rally']==rally]) for rally in data.rally.unique()]
plt.title("Histogram of Rally Length")
plt.xlabel("Rally Length")
plt.ylabel("Count")
plt.hist(rally_length, bins=max(rally_length))
plt.show()
```

```python
sum([1 for rl in rally_length if rl<=2])/len(rally_length)
```

```python
target_players = ['Anders ANTONSEN',
 'Anthony Sinisuka GINTING',
 'CHOU Tien Chen',
 'Jonatan CHRISTIE',
 'Kento MOMOTA',
 'NG Ka Long Angus',
 'Viktor AXELSEN']
```

```python
player2rc = {player: 0 for player in target_players}
for rally in data.rally.unique():
    if data.loc[data['rally']==rally].iloc[0]['name_A'] in player2rc:
        player2rc[data.loc[data['rally']==rally].iloc[0]['name_A']] += 1
    if data.loc[data['rally']==rally].iloc[0]['name_B'] in player2rc:
        player2rc[data.loc[data['rally']==rally].iloc[0]['name_B']] += 1
player2rc
```

```python
shot_types = data['type'].unique()
```

```python
player2shots = {player: {shot:0 for shot in shot_types} for player in target_players}
```

```python
for idx ,row in data.iterrows():
    player = row['name_'+row['player']]
    if player in player2shots:
        player2shots[player][row['type']] += 1
```

```python
#p2s = np.array([[player2shots[player][shot] for shot in shot_types if shot not in ['發長球', '發短球']] for player in player2shots])
p2s = np.array([[player2shots[player][shot] for shot in shot_types if shot] for player in player2shots])
```

```python
shot_types_en = ['short service', 'long', 'push', 'smash', 'return net',
                'rush', 'drive', 'net shot', 'lob', 'wrist smash', 'cross-court net shot',
                'passive drop', 'defensive drive', 'defensive lob', 'long service',
                'drop', 'back-court drive', 'unknown', 'driven flight']

idx2player = {0:'ANTONSEN', 1:'GINTING', 2:'CHOU',
       3:'CHRISTIE', 4:'MOMOTA', 5:'NG',
       6:'AXELSEN'}

idx2shot = {i: shot_types[i] for i in range(len(shot_types))}
shot2en = {shot_types[i]: shot_types_en[i] for i in range(len(shot_types))}
```

```python
p2s_norm = p2s/np.repeat(np.expand_dims(p2s.sum(axis=1), axis=1), p2s.shape[1], axis=1)
#p2s_norm_round = np.around(p2s_norm, decimals=3)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
df = pd.DataFrame(p2s_norm, [idx2player[i] for i in range(p2s.shape[0])], [shot2en[idx2shot[i]] for i in range(p2s.shape[1])])
sn.heatmap(df, annot=True, annot_kws={"size": 8}, cmap='GnBu', fmt='.2f', ax=ax, square=True)
b, t = ax.get_ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
ax.set_title("Players' Shot Types Distribution")
ax.set_ylim(b, t) # update the ylim(bottom, top) values
ax.xaxis.set_tick_params(rotation=90)
ax.yaxis.set_tick_params(rotation=0)
ax.set_xlabel('Shot Type')
ax.set_ylabel('Player')
for item in ([ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
ax.title.set_fontsize(16)
```

```python
kl_mtrx = np.array([[entropy(p2s_norm[i], qk=p2s_norm[j]) for j in range(p2s_norm.shape[0])] for i in range(p2s_norm.shape[0])])
#kl_mtrx = np.around(kl_mtrx, decimals=3)
```

```python
kl_mtrx
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
df = pd.DataFrame(kl_mtrx, [idx2player[i] for i in range(p2s.shape[0])], [idx2player[i] for i in range(p2s.shape[0])])
sn.heatmap(df, annot=True, annot_kws={"size": 10}, cmap='GnBu', fmt='0.2f', ax=ax, square=True)
b, t = ax.get_ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
ax.set_title("KL Divergence of Shot Types between Players")
ax.set_ylim(b, t) # update the ylim(bottom, top) values
ax.xaxis.set_tick_params(rotation=90)
ax.yaxis.set_tick_params(rotation=0)
ax.set_xlabel('Plyaer')
ax.set_ylabel('Player')
for item in ([ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
ax.title.set_fontsize(16)
```
